```python
#split into training and testing data
from sklearn.model_selection import train_test_split

X = data.drop(["median_house_value"],axis = 1) #axis = 1 because we are dropping a column
Y = data.median_house_value
```

```python
#total_rooms, total_bedrooms, population, households are all skewed data. Applying log function to normalize it
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
train_data["households"] = np.log(train_data["households"] + 1)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity))
```

```python
train_data = train_data.drop(["ocean_proximity"], axis = 1)
```

```python
#feature engineering

#number of bedrooms per room
train_data["bedroom_ratio"] = train_data["total_bedrooms"]/train_data["total_rooms"]
#number of rooms per household
train_data["household_rooms"] = train_data["total_rooms"]/train_data["households"]
```

```python
#heatmap again
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = train_data.drop(["median_house_value"], axis = 1)
Y = train_data.median_house_value
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
X_test
```

```python
test_data = X_test.join(Y_test)
```

```python
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.fit_transform(X_test)
```

```python
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```python
reg.score(X_test, Y_test)
```

```python
reg1 = LinearRegression()
reg1.fit(X_train_norm, Y_train)
```

```python
reg1.score(X_test_norm, Y_test)
```

```python
#RandomForestModel
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(X_train, Y_train)
```

```python
forest.score(X_test, Y_test) #higher score of 82.2% vs 69.8%
```

```python
#cross validation (divide training set into k folds; training all folds but one and test on that one; do this for all possible combinations; evaluate your parameters)

#parameter grid, grid search with cross validation
from sklearn.model_selection import GridSearchCV
#specify different hyperparameters in param_grid

forest = RandomForestRegressor()
#performs cross validation with all these combinations
param_grid = {
    "n_estimators": [100,200,300],
    "min_samples_split":[2,3,4,5,6],
    "max_features": [2,4,6,8],
    "max_leaf_nodes": [100,200,300,400],
    "max_depth": [None,4,8,10,12,14]
}

grid_search = GridSearchCV(forest, param_grid, cv= 5, scoring = "neg_mean_squared_error", return_train_score=True)
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_estimator_
```

```python
bestForest = grid_search.best_estimator_
```

```python
bestForest.score(X_test, Y_test) #hyperparameter tuning can make it worse but that is an anamoly. usually it improves the accuracy
```
